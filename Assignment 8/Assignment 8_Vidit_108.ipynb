9. Courses sorted by price from low to high.

```python
df[['course_title', 'price']].sort_values('price')
```

10. Course Ttitles sorted by the number of reviews in desccending order.

```python
df.sort_values('num_reviews', ascending = False)[['course_title', 'num_reviews']]
```

11. Course titles sorted by number of lectures in descending order.

```python
df.sort_values('num_lectures', ascending = False)[['course_title', 'num_lectures']]
```

12. Course title sorted by rating from best to worst.

```python
df.sort_values('rating', ascending = False)[['course_title', 'rating']]
```

13. Courses soted by duration from least to most.

```python
df.sort_values('content_duration', ascending = False)[['course_title', 'content_duration']]
```

14. Different Subjects with their respected number of courses.

```python
df.subject.value_counts()
```

15. Numbe of free and paid courses.

```python
df['Free/Paid'].value_counts()
```

16. pie chart of free to paid courses.

```python
fig = plt.figure(figsize=(2,2), dpi=300, facecolor="w")
price = df['Free/Paid'].value_counts()
plt.pie(price, labels = ['Paid', 'Free'], autopct="%.2f%%")
plt.title("Pie Chart Showing Distibution of Free to Paid courses.")
plt.show()
```

17. Count of ratings.

```python
plt.figure(figsize=(12,6), dpi=300)
plt.subplot(221)
fund= sns.countplot(x=df["rating"],palette="magma",edgecolor='black',saturation=0.50)
fund.set_xticklabels(fund.get_xticklabels(),fontsize=1.5)
plt.title("COUNT OF RATINGS",fontsize=15)
fund.set_xlabel("Counts", fontsize=12)
fund.set_ylabel("User Rating", fontsize=12)
plt.show()
```

18. General Trend and outliers.

```python
fig, axes = plt.subplots(2, 2,figsize=(10,6),dpi=300)
fig.tight_layout()
fig.suptitle('Box Plots')
sns.boxplot(x=df["rating"], ax=axes[0,0],color="lightgreen")
sns.boxplot(x=df["num_reviews"],ax=axes[0,1])
sns.boxplot(x=df["price"],ax=axes[1,0])
sns.boxplot(x=df["num_subscribers"],ax=axes[1,1], color="lightgreen")
plt.show()
```
